## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
df2 = pd.read_csv("/Users/Kazuhide/chss_2023_assignment/課題0705/Data/state-abbrevs.csv")
df3 = pd.read_csv("/Users/Kazuhide/chss_2023_assignment/課題0705/Data/state-areas.csv")
df4 = pd.read_csv("/Users/Kazuhide/chss_2023_assignment/課題0705/Data/state-population.csv")
````

````python
#データフレームの結合
##2010年&全年代で絞り込み
df5 = df4[(df4["year"] == 2010) & (df4["ages"] == "total")]
##df2とdf3の結合
df6 = pd.merge(df2,df3)
##df5をdf6と結合できる形に変更
df5_new = df5.rename(columns={'state/region': 'abbreviation'})
##全てを一つにまとめる
df7 = pd.merge(df5_new, df6, on='abbreviation')
display(df7)
````

````python
#2010年アメリカの各州の人口密度を計算
df8 = df7["population"] / df7["area (sq. mi)"]
df9 = pd.DataFrame(df8)
df9
````
