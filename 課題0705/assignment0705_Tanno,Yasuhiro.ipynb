## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
df_abbrevs = pd.read_csv("/Users/yass/Documents/VSCode/chss_2023_assignment/課題0705/Data/state-abbrevs.csv")
df_areas = pd.read_csv("/Users/yass/Documents/VSCode/chss_2023_assignment/課題0705/Data/state-areas.csv")
df_population = pd.read_csv("/Users/yass/Documents/VSCode/chss_2023_assignment/課題0705/Data/state-population.csv")

# 州の略称
# 州の面積（平方マイル）
# 州の人口（2010, total取り出し）＝＞略称で書かれている

df1 = df_population[df_population["ages"] == "total"] #df_populationからagesがtotalのものを取り出す
df2 = df1[df1["year"] == 2010] #df1からyearが2010のものを取り出す

df3 = pd.merge(df_abbrevs, df_areas) # df_abbrevsとdf_areasをstateを軸に結合
df4 = pd.merge(df2, df3, left_on="state/region", right_on="abbreviation") # df2とdf3をstate/region=abbreviationとして結合

def dencity(row):
    return row['population']/row['area (sq. mi)'] #行を引数に取り、populationラベルの値/area (sq. mi)ラベルの値　を返す関数dencityを定義

df5 = df4

df5["dencity"] = df5.apply(dencity, axis=1) #関数densityの値の行を作成
df5
````
