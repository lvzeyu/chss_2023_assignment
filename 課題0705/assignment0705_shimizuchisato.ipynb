pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
df1 = pd.read_csv("Data/state-abbrevs.csv")
df2 = pd.read_csv("Data/state-areas.csv")
df3 = pd.read_csv("Data/state-population.csv")

df1
df2
display(df3)

df4 = pd.merge(df1,df2,on="state",how="outer")
display(df4)
````

````python
df5 = df3[df3["year"]==2010]
df6 = df5[df5["ages"]=="total"]
df6 = df6.rename(columns={'state/region':'abbreviation'})
df7 = pd.merge(df4,df6,on="abbreviation",how="outer")
display(df7)
````

````python
df7["pd"]=df7["population"]/df7["area (sq. mi)"]
df7
````
