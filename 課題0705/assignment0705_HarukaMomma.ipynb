## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
import pandas as pd
import os
````

````python
#データ取り出し
abbrevs=pd.read_csv(r'C:\Users\Harekaze\Desktop\chss_2023_assignment\課題0705\Data\state-abbrevs.csv')
abbrevs
````

````python
#データ取り出し
areas=pd.read_csv(r'C:\Users\Harekaze\Desktop\chss_2023_assignment\課題0705\Data\state-areas.csv')
areas
````

````python
#データ取り出し　面積のみ
acreage=areas[["area (sq. mi)"]]
acreage
````

````python
#データ取り出し
popu=pd.read_csv(r'C:\Users\Harekaze\Desktop\chss_2023_assignment\課題0705\Data\state-population.csv')
popu
````

````python
abbare_data = pd.concat([abbrevs,areas], axis=1)
abbare_data1=abbare_data.rename(columns={'abbreviation': 'state/region'})
abbare_data2=abbare_data1.drop("state", axis=1)
abbare_data2
````

````python
#abbrevation area population　すべてのデータを統合

newdata=pd.merge(popu, abbare_data2, on="state/region")
newdata
````

````python
#2010の人口密度

popu_2010 =newdata.loc[newdata["year"]==2010,["population"]]
popu_2010

for i in newdata:
    if popu_2010.index[i]==newdata["area (sq. mi)"][i]:
        print(popu_2010/newdata["area (sq. mi)"])
    else:
        pass
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
###state/region
popu_strg=newdata.groupby("state/region")["population"].mean()
popu_strg

for i in newdata:
    if popu_strg[i]==newdata["area (sq. mi)"][i]:
        print(popu_strg/newdata["area (sq. mi)"])
    else:
        pass
````

````python
###ages
popu_ages=newdata.groupby("ages")["population"].mean()/newdata["area (sq. mi)"]

for i in newdata:
    if popu_ages[i]==newdata["area (sq. mi)"][i]:
        print(popu_ages/newdata["area (sq. mi)"])
    else:
        pass
````

````python
###year
popu_year=newdata.groupby("year")["population"].mean()/newdata["area (sq. mi)"]

for i in newdata:
    if popu_year[i]==newdata["area (sq. mi)"][i]:
        print(popu_year/newdata["area (sq. mi)"])
    else:
        pass
````
