## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
df1=pd.read_csv("/Users/yamazakimomoko/chss_2023_assignment/課題0705/Data/state-abbrevs.csv")
df1
````

````python
df2=pd.read_csv("/Users/yamazakimomoko/chss_2023_assignment/課題0705/Data/state-areas.csv")
df2
````

````python
df3=pd.read_csv("/Users/yamazakimomoko/chss_2023_assignment/課題0705/Data/state-population.csv")
df3
````

````python
df_state_pop = df3[df3['year'] == 2010]
df_state_pop = df_state_pop[df_state_pop['ages'] == 'total']

df_merged = pd.merge(df1, df2, on='state')
df_merged = pd.merge(df_merged, df_state_pop, left_on='abbreviation', right_on='state/region')

df_merged['population_density'] = df_merged['population'] / df_merged['area (sq. mi)']

print(df_merged[['state', 'population_density']])
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
df_grouped = df_merged.groupby(['state/region', 'ages', 'year']).sum()
df_grouped['population_density'] = df_grouped['population'] / df_grouped['area (sq. mi)']

print(df_grouped[['population_density']])
````
