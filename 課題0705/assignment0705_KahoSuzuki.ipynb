## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
abbrevs=pd.read_csv("state-abbrevs.csv")
areas=pd.read_csv("state-areas.csv")
population=pd.read_csv("state-population.csv")
````

````python
population2010=population[(population["ages"]=="total")&(population["year"]==2010)]
population2010
````

````python
df=pd.merge(population2010,abbrevs,left_on="state/region",right_on="abbreviation")
df=pd.merge(df,areas).drop(["state/region","year","ages"],axis=1)
df
````

````python
df["population_density"]=df["population"].div(df["area (sq. mi)"])
df=df.reindex(["state","abbreviation","area (sq. mi)","population","population_density"],axis=1)
df
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
df2=pd.merge(population,abbrevs,left_on="state/region",right_on="abbreviation")
df2=pd.merge(df2,areas).drop(["state","abbreviation"],axis=1)
df2=df2.reindex(["state/region","ages","year","area (sq. mi)","population"],axis=1)
df3=df2.groupby(["state/region","ages","year"]).sum()
df3["population_density"]=df3["population"].div(df3["area (sq. mi)"])
df3
````
