## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
import pandas as pd

df1 = pd.read_csv("C:\\Users\\makot\\test2\\chss_2023_assignment\\課題0705\\Data\\state-abbrevs.csv")
df2 = pd.read_csv("C:\\Users\\makot\\test2\\chss_2023_assignment\\課題0705\\Data\\state-areas.csv")
df3 = pd.read_csv("C:\\Users\\makot\\test2\\chss_2023_assignment\\課題0705\\Data\\state-population.csv")
````

````python
list_pop_den = []
list_abb = []
lib_pop_den = {}

df_a = pd.merge(df1, df2)
df_b = df3[(df3["ages"]=="total") & (df3["year"]==2010)]

for i in range(len(df_a)):
    for j in range(len(df3)):
        try:
            if df_b["state/region"][j] == df_a["abbreviation"][i]:
                list_pop_den.append(df_b["population"][j]/df_a["area (sq. mi)"][i])
        except:
            pass

for i in range(len(df_a)):
    list_abb.append(df_a["abbreviation"][i])

for i in range(len(list_pop_den)):
    lib_pop_den[list_abb[i]] = list_pop_den[i]

lib_pop_den
````

````python
df_a = pd.merge(df1, df2)
df_b = df3[(df3["ages"]=="total") & (df3["year"]==2010)]
df_c = pd.merge(df_a, df_b, left_on="abbreviation", right_on = "state/region")

def pop_den(x):
    return x["population"]/x["area (sq. mi)"]

df_c["pop/den"] = df_c.apply(pop_den, axis=1)
df_c
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
df_a = pd.merge(df1, df2)

df_d = pd.merge(df_a, df3, left_on="abbreviation", right_on = "state/region")

def pop_den(x):
    return x["population"]/x["area (sq. mi)"]
````

````python
df_d.groupby("state/region").apply(pop_den)
````

````python
df_d.groupby("ages").apply(pop_den)
````

````python
df_d.groupby("year").apply(pop_den)
````
