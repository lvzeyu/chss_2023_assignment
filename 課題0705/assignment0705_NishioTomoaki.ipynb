## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
df1 = pd.read_csv('Data/state-abbrevs.csv')
df2 = pd.read_csv('Data/state-areas.csv')
df3 = pd.read_csv('Data/state-population.csv')
````

````python
##2010年, 各州,  人口密度（人口/面積）
##面積はdf2, 人口・各州・時点はdf3
##USAを削除
df4 = df3[(df3["ages"] == "total") & (df3["year"] == 2010), ignore_index()]
````

````python
##PRとUSAいらない
df4 = df4.drop(2490)
df4 = df4.drop(2539)
````

````python
##51行目プエルトリコいらない
df2 = df2.drop(51)
````

````python
df4_new = df4.rename(columns={"state/region": "abbreviation"})
````

````python
df5 = pd.merge(df1, df2, on="state")
````

````python
df6 = pd.merge(df5, df4_new, on = "abbreviation")
````

````python
df6
````

````python
df6["density"] = df6["population"] / df6["area (sq. mi)"]
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
df3
````

````python
##PRとUSAいらない
df7 = df3[(df3["state/region"] != "USA") & (df3["state/region"] != "PR")]
````

````python
df7
````

````python
df7_rename = df7.rename(columns={"state/region": "abbreviation"})
````

````python
df8 = pd.merge(df7_rename, df1, on = "abbreviation")
df9 = pd.merge(df8, df2, on="state")
````

````python
df9["density"] = df9["population"] / df9["area (sq. mi)"]
````

````python
##州、年齢、年代でグループを分けて、人口密度を求めてもグループ分けした意味ない？
df9.groupby(["abbreviation", "ages", "year"])["density"].agg("mean")
````
