## 課題１

- ```NumPy```を使って、最小二乗法で重回帰分析の係数を推定する関数を作成しよう。
- 与えられたデータで実装して動作を確認せよ。

````python
def get_PartialRegressionCoefficient(x, y):
    import numpy as np

    # デザイン行列
    design = np.insert(x, 0, np.array([1 for _ in range(len(x))]), axis = 1)
    print(np.shape(design))

    # 行列計算
    xtx = np.dot(design.T, design)
    xtxxt = np.dot(np.linalg.inv(xtx), design.T)
    xtxxty = np.dot(xtxxt, y)

    return(xtxxty)
````

````python
# このコードを実行し、データの読み込み
from sklearn.datasets import load_iris
iris = load_iris()
# Irisデータセットをロード
iris = load_iris()
X = iris.data  # 説明変数のデータ
Y = iris.data[:, 1]  # 目的変数のデータ（がく片の幅）

# 説明変数の行列のなかから目的変数の行を削除
X = np.delete(X, 1, axis = 1)
````

````python
# 係数の算出
get_PartialRegressionCoefficient(X, Y)
````

# 課題2

````python
class Vector:
    import numpy as np
    def __init__(self, v1, v2):
        self.v1 = v1
        self.v2 = v2
    
    def size_check(self):
        if np.shape(self.v1) == np.shape(self.v2):
            return("same size")
        else:
            return("different size")
    
    def projection(self):
        return(
            np.dot(np.dot(self.v1, self.v2) / np.dot(self.v2, self.v2)) , self.v2
        )
    
    def cos_sim(self):
        v1_norm = np.linalg.norm(self.v1)
        v2_norm = np.linalg.norm(self.v2)
        dot = np.dot(self.v1, self.v2)
        return(
            dot / (v1_norm*v2_norm)
        )        
````

### 確認，サイズ

````python
import numpy as np
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

example = Vector(v1, v2)

# サイズの大きさの確認
example.size_check()
````

### 射影

````python
example.projection()
````

### コサイン類似度

````python
example.cos_sim()
````

```NumPy```を使用してランダムウォーク（random walk）をシミュレーションするプログラムを実装しよう。

````python
def randomwalk(n):
    import numpy as np
    import matplotlib.pyplot as plt
    x = [0]
    y = [0]
    for _ in range(n):
        x.append(x[_] + np.random.normal(0, .1))
        y.append(y[_] + np.random.normal(0, .1))

    # 描画
    plt.plot(x, y)
    plt.show()
````

````python
randomwalk(1000)
````
